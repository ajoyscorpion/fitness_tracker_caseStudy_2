```python
import numpy as np
```

```python
sleep_condition = [
    (fitness_dailyActivity_sleep['TotalMinutesAsleep'] >360) & (fitness_dailyActivity_sleep['TotalMinutesAsleep'] <= 480),(fitness_dailyActivity_sleep['TotalMinutesAsleep']>480)
]
categories_sleep = ['6h-8h', '> 8h']
```

```python
calorie_conditions = [
    (fitness_dailyActivity_sleep['Calories'] > 1500) & (fitness_dailyActivity_sleep['Calories'] <= 2500),
    (fitness_dailyActivity_sleep['Calories'] > 2500)
]
categories_calorie = ['1.5k-2.5k', '> 2.5k']
```

```python
distance_conditions = [
    (fitness_dailyActivity_sleep['TotalDistance'] > 5) & (fitness_dailyActivity_sleep['TotalDistance'] <= 10),
    (fitness_dailyActivity_sleep['TotalDistance'] > 10)
]
categories_distance = ['5km-10km', '> 10km']
```

```python
daily_activity_sleep = fitness_dailyActivity_sleep.assign(
    sleep_categories=np.select(sleep_condition,categories_sleep, default='< 6h'),
    calorie_categories=np.select(calorie_conditions,categories_calorie, default='< 1.5k'),
    distance_categories=np.select(distance_conditions,categories_distance, default='<5km')
)
```

```python
daily_activity_sleep
```

Visualisation

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='distance_categories',y='Calories',data=daily_activity_sleep,palette='viridis')

plt.xlabel('Distance Categories')
plt.ylabel('Calories Burnt')
plt.title('Correlation between Distance and Calories Burnt')

plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='sleep_categories', y='Calories', hue='distance_categories', data=daily_activity_sleep, palette='viridis')

plt.xlabel('Sleep Categories')
plt.ylabel('Calories Burnt')
plt.title('Correlation between Sleep, Calories Burnt, and Distance Categories')

plt.show()
```
